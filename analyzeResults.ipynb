```python
import spacy
import pandas as pd
import json
from plotly import express as px
```

```python
with open('pg-stats.jsonl') as f:
    data = f.read()
```

```python
lines = []
for line in data.split('\n'):
    print(line)
    lines.append(json.loads(line))
```

```python
df = pd.DataFrame(lines)
```

```python
df['fn'] = df['fn'].apply(lambda x: x[8:])
```

```python
df['year'] = df['fn'].str.slice(1,5)
```

```python
df.sort_values(by='percDesc', ascending=False).head(30)
```

```python
df['percDesc'] = df['percDesc'].astype(float)
```

Remove outliers by cutting out the ones with more than 38% description, which is where the first fiction work appears (Conrad's *Lord Jim*)

```python
df = df[df['percDesc'] < 38]
```

```python
df[['year', 'percDesc']].groupby('year').mean().rolling(10).mean().plot(figsize=(8,6))
```
